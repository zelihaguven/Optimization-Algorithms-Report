```python
# PSO Visualization Codes 
import numpy as np
import plotly.graph_objects as go

# Fitness function
def f(x, y):
    return x**2 + y**2

# 2D grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# random particle positions
num_particles = 10
particles = np.random.uniform(-5,5,(num_particles,2))
particles_z = f(particles[:,0], particles[:,1])

# create figure 
fig = go.Figure()

# Surface plot
fig.add_trace(go.Surface(
    z=Z, x=X, y=Y,
    colorscale='Viridis',
    opacity=0.8,
    showscale=False
))

# Particles
fig.add_trace(go.Scatter3d(
    x=particles[:,0],
    y=particles[:,1],
    z=particles_z,
    mode='markers',
    marker=dict(size=6, color='red', opacity=0.9),
    name='Particles'
))

# Global Best (0,0)
fig.add_trace(go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    mode='text',
    text=[''],  # Unicode star
    textfont=dict(size=25, color='white'),
    name='Global Best'
))

# Layout
fig.update_layout(
    title='PSO Particle Positions',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Fitness',
        xaxis=dict(backgroundcolor="lightgrey"),
        yaxis=dict(backgroundcolor="lightgrey"),
        zaxis=dict(backgroundcolor="lightgrey")
    ),
    margin=dict(l=0, r=0, t=50, b=0)
)

fig.show()
```

##  Conceptual PSO Visualization Code
**Purpose:**  
- This code is a **conceptual illustration** of Particle Swarm Optimization (PSO).  
- It **does not implement the actual PSO algorithm**.  
- Its main goal is to **visualize the idea** that particles explore a fitness landscape and the global best solution exists.

**What it shows:**  
- **Red dots (particles):** Represent potential solutions randomly placed on the fitness landscape.  
- **White star (Global Best):** Represents the global optimum (fixed at `(0,0)`).  
- **Surface plot:** Shows the fitness function \( f(x, y) = x^2 + y^2 \).  
- There are **no velocity updates, no personal best (pbest), and no iteration-based movement**.  

**Usage in report:**  
- Good for explaining **PSO conceptually**.  
- Helps the reader **visualize particles exploring the search space** without introducing algorithmic complexity.  

---

```python
import numpy as np
import plotly.graph_objects as go

# fitness function 
def f(x, y):
    return x**2 + y**2  # Min at (0,0)

# PSO parameters
num_particles = 10
num_iterations = 50
w = 0.5   # inertia weight
c1 = 1.5  # cognitive coefficient
c2 = 1.5  # social coefficient

# Particle initialization
positions = np.random.uniform(-5, 5, (num_particles, 2))
velocities = np.random.uniform(-1, 1, (num_particles, 2))

# Personal best
pbest_pos = positions.copy()
pbest_val = np.array([f(x,y) for x,y in positions])

# Global best
gbest_index = np.argmin(pbest_val)
gbest_pos = pbest_pos[gbest_index].copy()


#  Fitness grid (for surface)

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)


# Animation frames

frames = []
for it in range(num_iterations):
    # Update velocities and positions
    r1, r2 = np.random.rand(num_particles, 1), np.random.rand(num_particles, 1)
    velocities = w*velocities + c1*r1*(pbest_pos - positions) + c2*r2*(gbest_pos - positions)
    positions += velocities

    # Update personal bests
    fitness = np.array([f(x,y) for x,y in positions])
    better_mask = fitness < pbest_val
    pbest_val[better_mask] = fitness[better_mask]
    pbest_pos[better_mask] = positions[better_mask]

    # Update global best
    gbest_index = np.argmin(pbest_val)
    gbest_pos = pbest_pos[gbest_index].copy()

    # Frame
    frames.append(go.Frame(
        data=[go.Scatter3d(
            x=positions[:,0],
            y=positions[:,1],
            z=[f(x,y) for x,y in positions],
            mode='markers',
            marker=dict(size=6, color='red'),
            name='Particles'
        ),
        go.Scatter3d(
            x=[gbest_pos[0]],
            y=[gbest_pos[1]],
            z=[f(gbest_pos[0], gbest_pos[1])],
            mode='text',
            text=[''],
            textfont=dict(size=25, color='white'),
            name='Global Best'
        )],
        name=f'Iteration {it+1}'
    ))


# 5. Initial Figure

fig = go.Figure(
    data=[go.Surface(z=Z, x=X, y=Y, colorscale='Viridis', opacity=0.8, showscale=False),
          frames[0].data[0], frames[0].data[1]],
    layout=go.Layout(
        title='PSO Optimization Animation',
        scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Fitness'),
        updatemenus=[dict(type='buttons',
                          showactive=False,
                          buttons=[dict(label='Play',
                                        method='animate',
                                        args=[None, {"frame": {"duration":200, "redraw":True},
                                                     "fromcurrent":True}]),
                                   dict(label='Pause',
                                        method='animate',
                                        args=[[None], {"frame": {"duration":0, "redraw":False},
                                                       "mode":"immediate"}])])])
)

# Add frames
fig.frames = frames

fig.show()
```

##  Animated / Real PSO Simulation Code
**Purpose:**  
- This code **actually simulates the PSO algorithm**.  
- Particles **move according to PSO rules**, updating velocities, personal best (pbest), and global best (gbest) iteratively.  
- It produces a **3D animated visualization** showing the convergence process.

**What it shows:**  
- **Red dots (particles):** Actively move in the search space according to PSO equations.  
- **White star (Global Best):** Moves when a better global solution is found.  
- **Surface plot:** Shows the fitness landscape \( f(x, y) = x^2 + y^2 \).  
- **Animation frames:** Show each iteration, allowing one to **see particles converge towards the global optimum**.  

**Usage in report:**  
- Excellent for demonstrating **how PSO actually works**.  
- Shows **particle movement, search dynamics, and convergence**, providing a deeper understanding than the conceptual plot.
